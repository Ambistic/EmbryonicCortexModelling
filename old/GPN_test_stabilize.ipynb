```python
from gpn3 import GrowingPlanarNetwork
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
random.seed(0)
np.random.seed(0)
gpn = GrowingPlanarNetwork()
gpn.init_square(6)
```

```python
def p_dupl():
    return np.random.random() > 0.4
```

```python
# gpn.show_all()
```

```python
gpn.D.nodes
```

```python
for i in range(100):
    node = random.choice(list(gpn.G.nodes))
    if p_dupl():
        print("Dup", node)
        gpn.duplicate_node(node)
    else:
        print("Rem", node)
        gpn.remove_node(node)
    gpn.check_all()
    couple_ngb = [(len(gpn.ngb(x)), x) for x in gpn.G.nodes if len(gpn.ngb(x)) > 5]
    print(couple_ngb)
```
